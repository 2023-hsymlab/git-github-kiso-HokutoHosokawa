```python
csv_mat = pd.read_csv("data_j.csv", usecols=[1,2])
print(csv_mat)
```

```python
csv_mat2 = csv_mat.T.values.tolist()
print(csv_mat2[1])
csv_mat2[1] = list(map(lambda val:val.replace('\u3000',' '), csv_mat2[1]))
print(csv_mat2)
```

```python
search_words = "ソ"
searched_list = [s for s in csv_mat2[1] if search_words in s]
print(searched_list)

searched_list_index = list(map(lambda val: csv_mat2[1].index(val),searched_list))
print(searched_list_index)
###上のインデックスの場所にある証券コードを引っ張ってきたい。
emp_search = []
emp_search.append(searched_list)
code_mat = []
print(emp_search)
for i in range(len(searched_list_index)):
    code_mat.append(csv_mat2[0][int(searched_list_index[i])])
emp_search.append(code_mat)
print(emp_search)
```
